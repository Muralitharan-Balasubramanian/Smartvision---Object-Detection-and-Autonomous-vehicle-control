# Conversion of BDD100K_train.json file to Yolo compatible .txt file

```python
import json
import os

#File path to BDD100K Train file
file_path_train = 'D:\\Smart vision\\BDD100K\\bdd100k_labels_release\\bdd100k\\labels\\bdd100k_labels_images_train.json'

#File path to label file
output_label_path = 'D:\\Smart vision\\BDD100K\\bdd100k_labels_release\\bdd100k\\labels\\train'

#Labelled_classs
classes = ['bike', 'bus', 'car', 'drivable area', 'lane', 'motor', 'person', 'rider', 'traffic light', 'traffic sign', 'train', 'truck']

os.makedirs(output_label_path, exist_ok = True)

#Usual dimensions of the BDD100K dataset images
image_width = 1280
image_height = 720

#Function that computes the Normalized co-ordinates of the center, width and height
def convert_box(box):
    x1, y1, x2, y2 = box['x1'], box['y1'], box['x2'], box['y2']
    x_center = (x1 + x2) / 2 / image_width
    y_center = (y1 + y2) / 2 / image_height
    width = (x2 - x1) / image_width
    height = (y2 - y1) / image_height
    return x_center, y_center, width, height

#Loading .json file
with open(file_path_train,"r") as f:
    data = json.load(f)

for i, item in enumerate(data):
    image_name  = item['name']
    labels = item['labels']

    #Create Output file .txt with the same base name as the real train image
    label_file_path = os.path.join(output_label_path, image_name.replace('.jpg','.txt'))

    with open(label_file_path,"w") as label_file:
        for obj in labels:
            category = obj.get('category')
            if category not in classes:
               continue # Skips if the class is unknown
            class_id = classes.index(category) 
            if 'box2d' not in obj:
               continue # Skips if there is no bounding boxes
            x , y, w, h = convert_box(obj['box2d']) 
            label_file.write(f"{class_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")

    if i % 1000 == 0:
        print(f"{i} images processed successfully...") # ~69000 Images processed

print("Conversion completed. YOLO labels saved to: ", output_label_path)
```

# To extract samples from the two sources

```python
import shutil
import os
from tqdm import tqdm

#Source Image directory
src_dirs = {
    'train1': r'D:\Smart vision\BDD100K\bdd100k\bdd100k\images\100k\train\trainA',
    'train2': r'D:\Smart vision\BDD100K\bdd100k\bdd100k\images\100k\train\trainB',
    'train_label': r'D:\Smart vision\BDD100K\bdd100k_labels_release\bdd100k\labels\train'
}

#Target Directory
tar_dirs = {
    'train_image': r'D:\SmartVision (Computer Vision)\Sample train-12000\images\train',
    'train_label': r'D:\SmartVision (Computer Vision)\Sample train-12000\labels\train',
}

os.makedirs(tar_dirs['train_image'], exist_ok=True)
os.makedirs(tar_dirs['train_label'], exist_ok=True)

#Sample Size 
sample_size = 1000

def extract_samples(sources, target_image, target_label, sample_size, dataset_type):
   #Extract sample function is used to extract samples from the sources into target directory
   total_copied = 0 
   img_file_copied = []
   lbl_file_copied = []

   #Fetching samples per source info to get the balanced images from the datasets
   samples_per_source = sample_size // len(sources)

   print(f"\nExtracting {sample_size} {dataset_type} samples ({len(sources)} sources)...")
    
   for i, source in enumerate(sources):
       # Getting list of image files
       image_files = [f for f in os.listdir(source) if f.lower().endswith(('jpg','png'))]

       sample_files = image_files[:samples_per_source]
       
       for img_file in tqdm(sample_files, desc= f'Processing {dataset_type} from source {i+1}'):
           src_img_path = os.path.join(source, img_file)
           
           base_name = os.path.splitext(img_file)[0]
           img_ext = os.path.splitext(img_file)[1]
           
           img_file_name = f'{base_name}{img_ext}'
           label_file_name = f'{base_name}.txt'

           src_lbl_path = os.path.join(src_dirs['train_label'], label_file_name)

           tar_img_path = os.path.join(target_image, img_file_name)
           tar_lbl_path = os.path.join(target_label, label_file_name)

           shutil.copy(src_img_path, tar_img_path)
           
           if os.path.exists(src_lbl_path):
             shutil.copy(src_lbl_path, tar_lbl_path)
             lbl_file_copied.append(label_file_name)
             total_copied += 1
           else:
             print(f"Warning: No label found for {img_file}")

           img_file_copied.append(img_file_name)
        
   return total_copied, img_file_copied, lbl_file_copied
   
# Process training set (combine samples from both train sources and fetch labels)
train_copied, train_img_files, train_lbl_files = extract_samples(
    [src_dirs['train1'], src_dirs['train2']],
    tar_dirs['train_image'],
    tar_dirs['train_label'],
    sample_size,
    'train'
)

# Print summary
print("\nOperation Summary:")
print(f"Total images copied: {len(train_img_files)}")
print(f"Images with matching labels: {train_copied}")
print(f"Images without labels: {len(train_img_files) - train_copied}")
```

# Convert .ppm to .txt file

```python
import os
from PIL import Image

input_dir = "D:\\SmartVision (Computer Vision)\\TrainIJCNN2013\\images\\train"
label_dir = "D:\\SmartVision (Computer Vision)\\TrainIJCNN2013\\labels"
os.makedirs(label_dir, exist_ok=True)

with open("D:\\SmartVision (Computer Vision)\\TrainIJCNN2013\\gt.txt", "r") as f:
    lines = f.readlines()

for line in lines:
    items = line.strip().split(";")
    image_file = items[0].replace(".ppm", ".jpg")
    xmin, ymin, xmax, ymax = map(int, items[1:5])
    class_id = int(items[5])

    img_path = os.path.join(input_dir, image_file)
    with Image.open(img_path) as img:
        W, H = img.size

    # Convert to YOLO format
    x_center = ((xmin + xmax) / 2) / W
    y_center = ((ymin + ymax) / 2) / H
    width = (xmax - xmin) / W
    height = (ymax - ymin) / H

    label_path = os.path.join(label_dir, image_file.replace(".jpg", ".txt"))
    with open(label_path, 'a') as out:
        out.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
```
